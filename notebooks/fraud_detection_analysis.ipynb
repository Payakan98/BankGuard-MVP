```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import numpy as np

DF = pd.read_csv('data/transactions_sample.csv', parse_dates=['timestamp'])
DF['hour'] = DF['timestamp'].dt.hour
DF['log_amount'] = np.log1p(DF['amount'])

X = DF[['log_amount','hour']].fillna(0)
scaler = StandardScaler()
Xs = scaler.fit_transform(X)
clf = IsolationForest(n_estimators=200, contamination=0.01, random_state=42)
clf.fit(Xs)

DF['anomaly'] = clf.predict(Xs)
print("Anomalies détectées :", (DF['anomaly'] == -1).sum())
```
